```python
#Vectorising corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
vect = CountVectorizer(binary=True)
```

```python
corpus=["NLP is awesome","I want to start an NLP company","Top unicorns use NLP"]
```

```python
vect.fit(corpus)
```

```python
vocab = vect.vocabulary_
```

```python
for key in sorted(vocab.keys()):
    print("{}:{}".format(key,vocab[key]))
```

```python
vect.transform(["I have an awesome NLP company worth 1 billion"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vect.transform(["NLP is an elective subject"]).toarray(),vect.transform(["NLP is attractive career now"]).toarray())
```

```python
print(similarity)
```
